### 1. Johdetaan projektiovektorin kaava geometrisesti. ###

Olkoon $a\neq 0$ suuntavektori ja $b$ vektori, jonka projektiota haluamme tarkastella suuntaan $a$.

Projektio on vektori, joka

* on samansuuntainen kuin $a$
* kuvaa kuinka suuri osa vektorista $b$ on suunnassa $a$.

Koska projektiovektori on samansuuntainen kuin $a$, sen täytyy olla muotoa

$$
\mathrm{proj}_a(b)=k a
$$

jollakin skalaarilla (k).

Seuraavaksi ratkaistaan (k).

Pistetulon avulla tiedetään, että vektorin (b) komponentin pituus suunnassa (a) on

$$
|b|\cos\theta
$$

missä $\theta$ on vektorien välinen kulma.

Pistetulon määritelmä antaa

$$
a\cdot b = |a||b|\cos\theta
$$

josta saadaan

$$
|b|\cos\theta = \frac{a\cdot b}{|a|}
$$

Tämä on projektion pituus suunnassa (a).

Koska yksikkövektori suuntaan (a) on

$$
\hat a = \frac{a}{|a|}
$$

saadaan projektiovektori kertomalla projektion pituus yksikkövektorilla:

$$
\mathrm{proj}_a(b) =
\left(
\frac{a\cdot b}{|a|}
\right)
\frac{a}{|a|}
$$

Yhdistämällä nimittäjät:

$$
\mathrm{proj}_a(b) =
\frac{a\cdot b}{|a|^2}a
$$

Koska

$$
|a|^2=a\cdot a
$$

saadaan tavallisin esitysmuoto:

$$
\boxed{
\mathrm{proj}_a(b) =
\frac{a\cdot b}{a\cdot a}a
}
$$

---

Tarkistetaan tämä projektissamme.

Meillä on

$$
a=(2,1), \qquad b=(4,7)
$$

Lasketaan pistetulot:

$$
a\cdot b =  2\cdot4+1\cdot7 = 15
$$

ja

$$
a\cdot a = 2^2+1^2 = 5
$$



Sijoitus kaavaan:

$$
\mathrm{proj}_a(b) = \frac{15}{5}(2,1)
$$

siis

$$
\mathrm{proj}_a(b) = 3(2,1)
$$

ja lopulta

$$
\boxed{
\mathrm{proj}_a(b) = (6,3)
}
$$

Täsmälleen sama vektori c=(6,3), jonka piirsimme kuvaan.


```python
# Kirjoitetaan projektiokaava Pythonille funktioksi ja lasketaan sillä esimerkkiprojektiomme

import numpy as np

def projektio(a, b):
    """
    Laskee vektorin b projektion suuntaan a.
    """
    a = np.array(a, dtype=float)
    b = np.array(b, dtype=float)

    return (np.dot(a, b) / np.dot(a, a)) * a


# Testi projektimme arvoilla
a = np.array([2, 1])
b = np.array([4, 7])

c = projektio(a, b)       # kutsuu funktiota

print("Projektio:", c)
```

### 2. Projektiomatriisi

Vektorin projisoinnit voidaan suorittaa myös **matriisioperaattorin** avulla.

Projektio-vektorin kaava oli

$$
\mathrm{proj}_a(b) =
\frac{a\cdot b}{a\cdot a}a
$$

Olkoon suuntavektori $a=( a_1 \ a_2$)  ja projisoitava vektori $b=( a_1 \ a_2)$.
Kirjoittamalla pistetulot auki ja sijoittamalla ne projektiokaavaan, voidaan johtaa projektiomatriisi $P$ suuntaan $a$
$$
\boxed{
P = \frac{1}{a_1^2 + a_2^2}
\begin{pmatrix}
a_1^2 & a_1 a_2 \\
a_1 a_2 & a_2^2
\end{pmatrix}
}
$$

Tällöin vektorin $b$ projektio vektorin $a$ suuntaan saadaan kertomalla $b$ matriisilla $P$:

$$
\mathrm{proj}_a(b)=Pb
$$


 ja siis 

$$
\boxed{
\operatorname{proj}_{a}(b) =
\frac{1}{a_1^2 + a_2^2}
\begin{pmatrix}
a_1^2 & a_1 a_2 \\
a_1 a_2 & a_2^2
\end{pmatrix}
\begin{pmatrix}
b_1 \\
b_2
\end{pmatrix}
}
$$


---
#### Projektiovektorin lasku 

Käytetään esimerkkivektoreita

$$
a=
\begin{pmatrix}
2\
1
\end{pmatrix}
$$

$$
b=
\begin{pmatrix}
4\
7
\end{pmatrix}
$$

Projektiomatriisi oli

$$
P=
\frac{1}{a_1^2+a_2^2}
\begin{pmatrix}
a_1^2 & a_1a_2\\
a_1a_2 & a_2^2
\end{pmatrix}
$$

Sijoitetaan komponentit

$$
a_1=2
$$

$$
a_2=1
$$

Saadaan

$$
P=
\frac{1}{2^2+1^2}
\begin{pmatrix}
2^2 & 2\cdot1\\
2\cdot1 & 1^2
\end{pmatrix} =
\frac{1}{5}
\begin{pmatrix}
4 & 2\\
2 & 1
\end{pmatrix}
$$

Projektio saadaan kertomalla vektori $b$ matriisilla $P$

$$
\operatorname{proj}_{a}(b)=Pb =
\frac{1}{5}
\begin{pmatrix}
4 & 2\\
2 & 1
\end{pmatrix}
\begin{pmatrix}
4\\
7
\end{pmatrix} =
\frac{1}{5}
\begin{pmatrix}
4\cdot4+2\cdot7\\
2\cdot4+1\cdot7
\end{pmatrix} =
\frac{1}{5}
\begin{pmatrix}
30\\
15
\end{pmatrix} =
\begin{pmatrix}
6\\
3
\end{pmatrix}
$$

Siis

$$
\boxed{
\operatorname{proj}_{a}(b)=
\begin{pmatrix}
6\\
3
\end{pmatrix}
}
$$

Tulos on sama kuin aiemmin geometrisesti ja pistetulokaavalla laskettu projektiovektori.

---

```python
###  Python-koodi: projektiomatriisi funktiona

import numpy as np

def projektio_matriisilla(a, b):
    """
    Laskee projektiomatriisin P ja
    vektorin b projektion suuntaan a.
    """

    a = np.array(a, dtype=float).reshape(2, 1)
    b = np.array(b, dtype=float).reshape(2, 1)

    # Projektiomatriisi
    P = (a @ a.T) / (a.T @ a)

    # Projektiovektori
    proj = P @ b

    print("Projektiomatriisi P:")
    print(P)

    print("\nProjektioveektori proj_a(b):")
    print(proj)

    return P, proj


# Esimerkki
a = [2, 1]
b = [4, 7]

P, proj = projektio_matriisilla(a, b)  # kutsuu funktiota 
```

### 3. Matriisit töihin"

**Ohjelmointiohje ChatGTP:lle**

- Koodaa funktio, joka pyörittää (rotaatiomatriisi) annettua vektoria b 5 asteeen välein 360 astetta ja tallentaa tulostuksena näin saadut vektorit listaksi.
- Koodaa toinen funktio, joka ottaa  syötteeksi edellä luodun listan ja projisoi ne projektiomatriisilla P annetun suuntavektorin a suuntaan ja laskee kunkin projektiovektorin normin. 
- Lopuksi koodi tulostaa graafisesti pyöritettyjen vektoreiden projektioiden normit astelukujensa funktiona.
  

---
**ChatGTP:n antaa suunnitelmakseen**

Koodi tekee kolme asiaa:

1. **Pyöritysfunktio**

  - pyörittää vektoria b rotaatiomatriisilla 5° välein välillä 0
∘
…360
∘


  - tallentaa saadut vektorit listaksi


2. **Projektiofunktio**
  - muodostaa projektiomatriisin annetusta suuntavektorista a

  - projisoi kaikki pyörityslistan vektorit suuntaan a

  - laskee jokaisen projektiovektorin normin

3.**Tulostus / visualisointi**
  - piirtää projektioiden normit kulman funktiona.


```python
import numpy as np
import matplotlib.pyplot as plt


# --------------------------------------------------
# 1. Vektorin pyöritys rotaatiomatriisilla
# --------------------------------------------------

def pyorita_vektori(b, step_deg=5):
    """
    Pyörittää vektoria b rotaatiomatriisilla
    0...360 astetta.

    Palauttaa:
        kulmat      : lista asteita
        vektorit    : lista pyöritetyistä vektoreista
    """

    b = np.array(b, dtype=float).reshape(2, 1)

    kulmat = []
    vektorit = []

    for theta_deg in range(0, 361, step_deg):

        theta = np.radians(theta_deg)

        # Rotaatiomatriisi
        R = np.array([
            [np.cos(theta), -np.sin(theta)],
            [np.sin(theta),  np.cos(theta)]
        ])

        b_rot = R @ b

        kulmat.append(theta_deg)
        vektorit.append(b_rot)

    return kulmat, vektorit


# --------------------------------------------------
# 2. Projektiot ja normit
# --------------------------------------------------

def projektio_normit(a, vektorilista):
    """
    Projisoi vektorit suuntaan a
    projektiomatriisilla ja laskee normit.
    """

    a = np.array(a, dtype=float).reshape(2, 1)

    # Projektiomatriisi
    P = (a @ a.T) / (a.T @ a)

    normit = []

    for v in vektorilista:

        proj = P @ v
        normi = np.linalg.norm(proj)

        normit.append(normi)

    return P, normit


# --------------------------------------------------
# 3. Esimerkki
# --------------------------------------------------

a = [2, 1]
b = [4, 7]

# Pyöritys
kulmat, vektorit = pyorita_vektori(b)

# Projektiot
P, normit = projektio_normit(a, vektorit)

print("Projektiomatriisi P:")
print(P)

# Kuvaaja
plt.figure(figsize=(8,5))
plt.plot(kulmat, normit, 'o-')

plt.xlabel("Kulma (astetta)")
plt.ylabel("Projektioiden normi")
plt.title("Pyöritettyjen vektoreiden projektioiden normit")
plt.grid(True)

plt.show()
```
